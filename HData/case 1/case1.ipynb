```python
import pandas as pd
```

```python
# Tabela dados recebidos (abas)

df_aba_lancamento = pd.read_excel('Dados Financeiros - Dados Recebidos.xlsx', sheet_name='LANCAMENTO')

df_aba_saldo_inicial = pd.read_excel('Dados Financeiros - Dados Recebidos.xlsx', sheet_name='SALDOS INICIAIS')

df_aba_depara_centro_custos = pd.read_excel('Dados Financeiros - Dados Recebidos.xlsx', sheet_name='DEPARA_CENTRO_CUSTOS')

df_aba_depara_contas_financ = pd.read_excel('Dados Financeiros - Dados Recebidos.xlsx', sheet_name='DEPARA_CONTAS_FINANC')
```

```python
# Tabela Estrutura HData

df_aba_fluxo_caixa = pd.read_excel('Dados Financeiros - Estrutura HData.xlsx', sheet_name='Fluxo Caixa - Planilha Padrao')

df_aba_dim_centro_custos = pd.read_excel('Dados Financeiros - Estrutura HData.xlsx', sheet_name='dim_centro_custos')

df_aba_conta_financeira = pd.read_excel('Dados Financeiros - Estrutura HData.xlsx', sheet_name='dim_conta_financeira')
```

```python
#Fazendo o enriquecimento

#depara contas financeiras
df1 = pd.merge(df_aba_lancamento, df_aba_depara_contas_financ, 
                          how='left', left_on='CONTA', right_on='ds_conta_financeira')

df1 = df1.drop(columns=['FORMA DE PAGAMENTO'])
df1 = df1.drop(columns=['ds_conta_financeira'])
df1 = df1.drop(columns=['VALOR PARCELA'])
df1 = df1.rename(columns={'VALOR REALIZADO':'VALOR PAGO'})
df1.fillna({'CONSIDERAÇÕES LANÇAMENTO': 'não mapeado'}, inplace=True)
df1.fillna({'NOTA FISCAL': 'não mapeado'}, inplace=True)
df1 = df1.dropna()
df1.columns = df1.columns.str.upper()

# Verifique se o merge ocorreu corretamente
df1
```

```python
#Exportando dados 
df1.to_excel('dados_exportados.xlsx', index=False)
```
