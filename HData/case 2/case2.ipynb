```python
import pandas as pd
```

```python
#Tabela atendimento (tratamentos)
df_atendimento = pd.read_csv('ATENDIMENTO.csv', delimiter=';')
df_atendimento.sort_values(by='NR_ATENDIMENTO', ascending=True, inplace=True)
df_atendimento = df_atendimento.dropna()
df_atendimento.loc[df_atendimento['IE_TIPO_ATENDIMENTO'] == 3, 'IE_TIPO_ATENDIMENTO'] = 'Pronto Atendimento'
df_atendimento.loc[df_atendimento['IE_CLINICA'] == 1, 'IE_CLINICA'] = 'Clinica Médica'
df_atendimento.loc[df_atendimento['IE_CLINICA'] == 4, 'IE_CLINICA'] = 'Pediatria'
df_atendimento.loc[df_atendimento['IE_CLINICA'] == 10, 'IE_CLINICA'] = 'Cirurgia Geral'

#Tabela PESSOA_FISICA_PAC (tratamentos)
df_pessoa_fisica_pac = pd.read_csv('PESSOA_FISICA_PAC.csv', delimiter=';', encoding='ISO-8859-1')
df_pessoa_fisica_pac = df_pessoa_fisica_pac.dropna()

#Tabela FILA_SENHA (tratamentos)
df_fila_senha = pd.read_csv('FILA_SENHA.csv', delimiter=';', encoding='ISO-8859-1')
df_fila_senha = df_fila_senha.dropna()

#Tabela MOTIVO_ALTA (tratamentos)
df_motivo_alta = pd.read_csv('MOTIVO_ALTA.csv', delimiter=';', encoding='ISO-8859-1')
df_motivo_alta = df_motivo_alta.dropna()

#Tabela PESSOA_FISICA_MEDICO (tratamentos)
df_pessoa_fisica_medico = pd.read_csv('PESSOA_FISICA_MEDICO.csv', delimiter=';', encoding='ISO-8859-1')
df_pessoa_fisica_medico = df_pessoa_fisica_medico.dropna()
```

```python
#Etapas monitoradas 

df_etapas_monitoradas = df_atendimento[['DT_ENTRADA', 'DT_RECEBIMENTO_SENHA','DT_FIM_TRIAGEM', 'DT_INICIO_ATENDIMENTO', 'IE_TIPO_ATENDIMENTO', 'IE_CLINICA', 'CD_MEDICO_RESP', 'DT_MEDICACAO', 'DT_ALTA']]
print(f"As etapas monitoradas são: ")

df_etapas_monitoradas.head()
```

```python
# Calcular o tempo de espera em minutos para cada etapa
df_etapas_monitoradas['tempo_espera_recebimento'] = (df_etapas_monitoradas['DT_RECEBIMENTO_SENHA'] - df_etapas_monitoradas['DT_ENTRADA']).dt.total_seconds() / 60
df_etapas_monitoradas['tempo_espera_triagem'] = (df_etapas_monitoradas['DT_FIM_TRIAGEM'] - df_etapas_monitoradas['DT_RECEBIMENTO_SENHA']).dt.total_seconds() / 60
df_etapas_monitoradas['tempo_espera_atendimento'] = (df_etapas_monitoradas['DT_INICIO_ATENDIMENTO'] - df_etapas_monitoradas['DT_FIM_TRIAGEM']).dt.total_seconds() / 60

# Adiciona uma coluna de prioridade com base no tempo total de espera
df_etapas_monitoradas['prioridade'] = df_etapas_monitoradas[['tempo_espera_recebimento', 'tempo_espera_triagem', 'tempo_espera_atendimento']].sum(axis=1)

# Ordena o DataFrame pela prioridade
df_etapas_monitoradas = df_etapas_monitoradas.sort_values(by='prioridade')

# Mostra os pacientes com maior prioridade
print(df_etapas_monitoradas[['NR_ATENDIMENTO', 'prioridade']])
```

```python
#Tempo medio de espera do paciente em cada etapa 

#calculando o tempo de espera entre as etapas 
df_etapas_monitoradas.loc[:, 'tempo_recebimento_senha'] = df_etapas_monitoradas['DT_RECEBIMENTO_SENHA'] - df_etapas_monitoradas['DT_ENTRADA']
df_etapas_monitoradas.loc[:, 'tempo_fim_triagem'] = df_etapas_monitoradas['DT_FIM_TRIAGEM'] - df_etapas_monitoradas['DT_RECEBIMENTO_SENHA']
df_etapas_monitoradas.loc[:, 'tempo_inicio_atendimento'] = df_etapas_monitoradas['DT_INICIO_ATENDIMENTO'] - df_etapas_monitoradas['DT_FIM_TRIAGEM']
df_etapas_monitoradas.loc[:, 'tempo_medicacao'] = df_etapas_monitoradas['DT_MEDICACAO'] - df_etapas_monitoradas['DT_INICIO_ATENDIMENTO']
df_etapas_monitoradas.loc[:, 'tempo_alta'] = df_etapas_monitoradas['DT_ALTA'] - df_etapas_monitoradas['DT_MEDICACAO']

#calculando  o tempo médio de espera entre as etapas
tempo_medio_recebimento_senha = df_etapas_monitoradas['tempo_recebimento_senha'].mean()
tempo_medio_fim_triagem = df_etapas_monitoradas['tempo_fim_triagem'].mean()
tempo_medio_inicio_atendimento = df_etapas_monitoradas['tempo_inicio_atendimento'].mean()
tempo_medio_medicacao = df_etapas_monitoradas['tempo_medicacao'].mean()
tempo_medio_alta = df_etapas_monitoradas['tempo_alta'].mean()

# Exibir os resultados
print(f"Tempo médio até o recebimento da senha: {tempo_medio_recebimento_senha}")
print(f"Tempo médio até o fim da triagem: {tempo_medio_fim_triagem}")
print(f"Tempo médio até o início do atendimento: {tempo_medio_inicio_atendimento}")
print(f"Tempo médio até a medicação: {tempo_medio_medicacao}")
print(f"Tempo médio até a alta: {tempo_medio_alta}")

```
